```python
def calculate_euclidean_distance(training_features, test_features):
    dist_arr = []
    for i in range(0,len(training_features)):

        tf = training_features.iloc[i,:-1]
        
        dist = np.sqrt(np.sum(np.square(test_features - tf)))
        dist_arr.append(dist)   
    return dist_arr
```

```python
#This function calculates the number of occurences of each of the k-nearest neighbors class and 
#retruns the class value that has maximum occurence
def calculate_occurence_of_eachclass(knearest_neighbors_class_list):
    count_per_inner_list = [inner_list.count(1) for inner_list in knearest_neighbors_class_list]
    total_count1 = sum(count_per_inner_list)
    count_per_inner_list = [inner_list.count(2) for inner_list in knearest_neighbors_class_list]
    total_count2 = sum(count_per_inner_list)
    
    if total_count1 > total_count2:
        return 1
    elif total_count2 > total_count1:
        return 2
    else:
        return 1
```

```python
#This function is the main algorithm which implements Version1 of modified HL-KNN algorithm. Takes the training data, 
#test data and predicts the class of the given test data.
#More description detailed in the report
def high_level_knn_modified (X_train,X_test,y_train):
    dist_arr = []
    predict_arr = []
    dist_arr1 = []
    arr1 = []
    arr12 = []
    predicted_testdata = pd.DataFrame()
    for i in range(0,len(X_test)):
        test = X_test.iloc[i,:-1]
        dist_arr = calculate_euclidean_distance(X_train, test)
        X_train['Distance']= dist_arr
        X_train['1']= y_train
        df = pd.DataFrame(X_train)
        df_sorted = df.sort_values(by='Distance')
        df_top7 = df_sorted.head(7)
        #print(df_top7)
        arr1.append(df_top7['1'].tolist())
        total1_count = calculate_occurence_of_eachclass(arr1)
        #new_df = df_top7[df_top7['1'] == total1_count].copy()
        
        X_train = X_train.drop('1',axis=1)
        if all(x == 1 for x in arr1):
            predict_arr.append(1) 
        elif all(x == 0 for x in arr1):
            predict_arr.append(0)
        else:
            for j in range(0,len(df_top7)):
                top7 = df_top7.iloc[j,:-1]
                dist_arr1 = calculate_euclidean_distance(X_train, top7)
                X_train['Distance']= dist_arr1
                X_train['1']= y_train
                df1 = pd.DataFrame(X_train)
                df1_sorted = df1.sort_values(by='Distance')
                df1_top7 = df1_sorted.head(7)
                arr1.append(df1_top7['1'].tolist())
                X_train = X_train.drop('1',axis=1)
            total_count = calculate_occurence_of_eachclass(arr1)
            predict_arr.append(total_count)
            arr1.clear()
            #arr12.clear()

    predicted_testdata['Female'] = X_test['Female']
    predicted_testdata['6.8'] = X_test['6.8']
    predicted_testdata['3.3'] = X_test['3.3']
    predicted_testdata['0.9'] = X_test['0.9']
    predicted_testdata['1'] = predict_arr

    return predicted_testdata
```

```python
def efficiency_test (predicted_testdata, X_test):
    counttt = 0
    for i in range (0,len(X_test)):
        if X_test.iloc[i,5] == predicted_test.iloc[i,4]:
            counttt = counttt +1
    percentage = (counttt/len(X_test)) * 100
    return percentage
```
